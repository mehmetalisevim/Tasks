```python
from scipy.stats import binom
repeating_number = [20, 20, 40]
probability_values = [0.2, 0.6, 0.6]
line_style = ['-', '--', ':']
color = ["darkblue", "darkgreen", "darkred"]
x = np.arange(0, 40)
fig, ax = plt.subplots(figsize=(12, 4), dpi=100)
for (n, b, ls, r) in zip(repeating_number, probability_values, line_style, color):
 dist = binom(n, b)
 y = dist.pmf(x)
 plt.plot(x, y, ls=ls, c=r, label=r'$b=%.1f,\ n=%i$' % (b, n))
plt.xlim(-1, 40)
plt.ylim(0, 0.25)
plt.ylabel('intensity', color="darkblue")
plt.legend()
plt.show()
```
