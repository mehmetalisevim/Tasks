**Practice** Practicing on modules.

```python
%%writefile example.py
def make_list_in_range(x,y):
  new_list=[]
  for a in range(x+1,y):
    new_list.append(a)
  return new_list

def make_list_of_even(x,y):

    new_list=[]
    for a in range(x+1,y):
      if a%2==0:
        new_list.append(a)
    return new_list
```

```python
try:
  make_list_in_range(1,34)
except Exception as e:
  print("'{}' error!".format(e))
```

**Practice** Write a function that specifies whether a number entered as a parameter is a prime number.

```python
prime_numbers = (int(input("enter a number: ")))
if prime_numbers > 1:
  for i in range(2, prime_numbers):
    if (prime_numbers % i) == 0:
      print("it is not a prime number")
      break
  else:
    print("it is a prime number")
else:
  print("it is a not prime number")
```
