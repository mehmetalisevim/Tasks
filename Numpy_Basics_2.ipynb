```python
x = np.ones((3,3))
print("Checkerboard pattern:")
x = np.zeros((8,8),dtype=int)
x[1::2,::2] = 1
x[::2,1::2] = 1
print(x)
```

```python
array_x = np.ones((10,10))
array_x[0] = 0
array_x[9] = 0
array_x
```

```python
my_array = np.arange(0,63,3)
new_array = my_array[::2]
new_array
```

```python
array_1 = np.array([10,20,30,40,50])
array_2 = np.array([5,10,20,34,48])
final_array = np.setxor1d(array_1,array_2)
final_array
```

```python
my_array = np.arange(0,160,3)
print("size: ", my_array.size)
print(my_array.itemsize)
print(my_array.nbytes)
```
