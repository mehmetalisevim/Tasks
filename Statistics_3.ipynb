**Practice:** Print a histogram and q-q table by using np.random.distribution_name. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(8,4), dpi = SS200)
sample = np.random.normal(32, 12, 123243)
plt.hist(sample, bins = 20, color ="lightcoral")
plt.axvline(np.mean(sample), color = "black", linestyle = "solid", linewidth = 2)
plt.axvline(np.mean(sample) + np.std(sample), color = "dimgray", linestyle = "dashed", linewidth = 2)
plt.axvline(np.mean(sample) - np.std(sample), color = "dimgray", linestyle = "dashed", linewidth = 2)
plt.show()
```

```python
normal_dist = np.random.normal(np.mean(sample), np.std(sample),123243)
normal_dist.sort()
sample.sort()
plt.figure(figsize=(8,4), dpi = 300)
plt.scatter(normal_dist,sample, color="green")
title_font = {"family": "arial", "color":"darkred","weight":"bold","size": 13}
axis_font = {"family":"arial","color":"black","weight":"bold","size":10}
plt.title("Sample Q-Q Plot", fontdict = title_font)
plt.xlabel("Normal Distribution", fontdict = axis_font)
plt.ylabel("Sample", fontdict = axis_font)
plt.plot([np.min(normal_dist), np.max(normal_dist)],[np.min(sample), np.max(sample)], c = "red")
plt.show()
```

**Conclusion:** The data that we created is normally distributed according to the histogram and the Q-Q plot.

**Practice:** Download google company stock quotes from 2015-2017 using yahoo finance website. Calculate the stock return using this data and discuss its distribution using the Q-Q chart.

```python
df = pd.read_csv("/content/drive/MyDrive/Datas/GOOG (3).csv")
df
```

```python
df["Yields"] = (df["High"]-df["Low"]) / df["Low"]
df
```

```python
df_2 = np.array(df["Yields"])
normal_dist_2 = np.random.normal(np.mean(df_2), np.std(df_2), 565)

normal_dist_2.sort()
df_2.sort()


plt.figure(figsize=(8,5), dpi = 300)

title_font = {"family": "arial", "color":"darkred","weight":"bold","size": 13}
axis_font = {"family":"arial","color":"darkblue","weight":"bold","size":10}
plt.title("2015-2017 Google Yields Q-Q Chart ", fontdict = title_font)
plt.xlabel("Normal Distribution", fontdict = axis_font)
plt.ylabel("Google", fontdict = axis_font)
plt.scatter(normal_dist_2, df_2)
plt.plot([np.min(normal_dist_2), np.max(normal_dist_2)],[np.min(df_2), np.max(df_2)], c = "red")
plt.show()
```

**Conclusions:** The data of Google Stock returns between 2015 and 2017 are not normally distributed. They are poisitively skewed according to the Q-Q plot. It means that the arithmetic mean of data is greater than the median and the median is greater than the mode.
