```python
country_code = ["fr","ge","us","uk","cn"]
countries = pd.DataFrame(index=country_code)
countries.index.name = "Code"
countries
```

```python
countries["countries"] = ["france","germany","usa","united kingdom","china"]
countries["population"] = [61083916, 82400996, 301139947, 60776238, 1318683096]
countries["lifetime"] = [80.657, 79.406, 78.242 ,79.425, 72.961]
countries["total income"] = [2657213249384, 3439953462907, 14477635000000,3074359743897, 3552182311652]
countries['continent'] = ['europe', 'europe', 'america', 'europe', 'asia']
countries["per capita income"] = (countries["total income"] / countries["population"]).astype(int)
countries
```
