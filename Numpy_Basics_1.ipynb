**Practice** Write a function that returns an array of 1's on the sides and 0's on the inside, depending on the number of rows and columns entered as a parameter.

```python
x_axis = int(input("please enter a x axis: "))
y_axis = int(input("please enter a y axis: "))
def func(x,y):
  new_array = np.zeros((x,y))
  new_array[0] = 1
  new_array[x-1] = 1
  new_array[:,0] = 1
  new_array[:,y-1] = 1
  return(new_array)
print(func(x_axis,y_axis))
```
