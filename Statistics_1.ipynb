```python
import matplotlib.pyplot as plt

pop=pd.DataFrame()

pop["low_var"] = np.random.normal(60,10,1000)
pop["high_var"] = np.random.normal(60,100,1000)

pop.hist(layout=(2,1), sharex=True)
plt.show()

print(pop.max())
print(pop.min())
```

```python
sample = pd.DataFrame()
sample["low_var"] = np.random.choice(pop["low_var"], 100)
sample["high_var"] = np.random.choice(pop["high_var"], 100)

sample.hist()
plt.show()

print(sample.mean())
print(sample.std(ddof=1))
```

**Practice:** Analyze the central tendency and distribution measurements through 3 different data you will produce using the NumPy library.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st 

import warnings
warnings.filterwarnings("ignore")

my_data = np.random.normal(10,8,100)
my_data
```

```python
print("Mean. ", np.mean(my_data)) 
print("Median: ", np.median(my_data))
print("Minimum Value: ", np.amin(my_data))
print("Maximum Value: ", np.amax(my_data))
print("Range of Maximum and Minimum: ", np.ptp(my_data))
print("Variance: ", np.var(my_data))
print("Standard deviation: ", np.std(my_data))
```
