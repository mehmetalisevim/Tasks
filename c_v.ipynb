```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import warnings
warnings.filterwarnings("ignore")

cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

print(len(X), len(y))
print("The dataset has {} rows..".format(len(X)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("The train dataset has {} rows ".format(len(X_train)))
print("The train dataset has {} targets ".format(len(y_train)))
print("The test dataset has {} rows ".format(len(X_test)))
print("The test dataset has {} targets ".format(len(y_test)))
```

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
sns.countplot(y)
plt.title("All datas")
plt.subplot(132)
sns.countplot(y_train)
plt.title("the train dataset")
plt.subplot(133)
sns.countplot(y_test)
plt.title("the test dataset")
plt.show()

print("All dataset 0 per cents : %{:.0f}".format(len(y[y==0])/len(y)*100))
print("the test data 0 per cents : %{:.0f}".format(len(y_test[y_test==0])/len(y_test)*100))
print("the train data 0 per cents : %{:.0f}".format(len(y_train[y_train == 0])/len(y_train)*100))
```
