**Practice** Write a number guessing game.

```python
guess=31
counter=1
x = input("guess a number:")
while not x.isdigit():
  x=input("guess a number:")

while int(x)!=guess:
  if guess>int(x):
    print("wrong number please guess upper")
  else:
    print("wrong number please guess lower")
  x=input("guess a number")
  while not x.isdigit():
    x=input("guess a number")
  counter+=1
print("You guessed correctly after %d trials." %counter) 
```

**Practice** Write a code in order to convert the Fahrenheit to the Celcius. 


```python
attempts = 5
while attempts > 0:
  attempts -= 1
  x=input("enter a tempereture as fahrenheit: ")
  c= 5/9*(int(x)-32)
  if attempts == 0:
    break
  print(c,"Celcius")
```

**Practice** Create a list consist of fibonacci numbers from 1 to 50.

```python
x=0
y=1
i=2
fibonacci_list =[x,y]
while i<50:
  fibonacci=x+y
  fibonacci_list.append(fibonacci)
  x=y
  y=fibonacci
  i+=1
print(fibonacci_list)
```
