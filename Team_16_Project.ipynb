### Bar plot

```python
# Bar plot representation of the gender distribution

plt.figure(figsize=(8,8))
plt.bar(df["Gender"].value_counts().index ,df["Gender"].value_counts(), color=["b","r"])

plt.title("Gender Distribution")
plt.ylabel("Number of students")

plt.show()
```

```python
# Bar plot representation of the education distribution

plt.figure(figsize=(8,8))
plt.bar(df["Education"].value_counts().index ,df["Education"].value_counts(), color=["b","r"])

plt.title("Education Distribution")
plt.ylabel("Number of students")

plt.show()
```

### Box plot

```python
df.drop(["Education"], axis=1, inplace=True)
df_female = df.loc[lambda x: df["Gender"] == "Female"]
df_male = df.loc[lambda x: df["Gender"] == "Male"]
df_male.drop(["Gender"], axis=1, inplace=True)
df_female.drop(["Gender"], axis=1, inplace=True)

plot_male = [df_male["Age"], df_male["Weight"], df_male["Height"]]
plot_female = [df_female["Age"], df_female["Weight"], df_female["Height"]]

plt.figure(figsize=(10,8), dpi=80)

plt.subplot(1,2,1)
plt.boxplot(plot_female, labels=("Age","Weight","Height"), boxprops={
'color': "blue"}, medianprops={"color":"green"})
plt.ylim(0,200)
plt.title("Boxplot of the Female Students")
plt.grid()

plt.subplot(1,2,2)
plt.boxplot(plot_male, labels=("Age","Weight","Height"), boxprops={
'color': "blue"}, medianprops={"color":"green"})
plt.title("Boxplot of the Male Students")
plt.ylim(0,200)
plt.grid()

plt.show()
```

### Scatter plot

```python
groups = df.groupby("Gender")
cdict = {"Female": 'red', "Male": 'blue'}

plt.figure(figsize=(7,6), dpi=100)

for name, group in groups:
    plt.scatter(group["Weight"], group["Height"], c=cdict[name], label=name)

plt.title("Attribute Values of Students")
plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.legend()
plt.show()
```
